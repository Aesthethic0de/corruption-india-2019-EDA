```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
corruption = pd.read_csv("Corruption 2019.csv")
```

```python
corruption.columns
```

```python
corruption 
```

```python
corruption.drop(["Category", "Unnamed: 11", "Unnamed: 12", "Unnamed: 13", "Unnamed: 14"], axis=1, inplace=True)
```

```python
corruption
```

```python
sns.pairplot(corruption)
```

```python
totalcase = corruption.sort_values("Total cases", ascending = False)
```

```python
totalcase
```

```python
plt.figure(figsize=(20,14))
sns.pointplot(data=totalcase, x='State/UT ', y='Total cases', color='red');
plt.title("Total Case of Corruption by state")
plt.tick_params(axis='x', which='major', labelsize=15, rotation=90)
plt.tick_params(axis='y', which='major', labelsize=15)
```

##### Rajasthan got the highest no of corruption case. = 2034

```python
chargesheet = corruption.sort_values("Cases Charge-sheeted", ascending = False)
```

```python
# what is chargesheet?
# A chargesheet is prepared after First Information Reports,
# and charges an individual for the crimes specified in those FIR. 
# Once the charge sheet has been submitted to a court of law, 
# the court decides as to who among the accused has sufficient 
# prima facie evidence against them to be put on trial.
```

```python
chargesheet
```

```python
plt.figure(figsize=(20,12))
plt.xticks(rotation=90)
sns.pointplot(x='State/UT ', y='Cases Charge-sheeted', data=chargesheet, color='red');
```

#### Maharashtra got the highest number of Chargesheet among other State till 2019.

```python
pending = corruption.sort_values("Cases Pending Investigation at End of the Year", ascending = False)
```

```python
plt.figure(figsize=(20,12))
plt.xticks(rotation=90)
sns.pointplot(x='State/UT ', y='Cases Pending Investigation at End of the Year', data=pending, color='red');
```

#### Rajasthan has the Highest Number of pending Investigation. 

```python
previous = corruption.sort_values("Cases Pending Investigation from Previous Year", ascending = False)
```

```python
plt.figure(figsize=(20,12))
plt.xticks(rotation=90)
sns.pointplot(x='State/UT ', y='Cases Pending Investigation from Previous Year', data=previous, color='red');
```

#### Rajasthan has the Highest Pending Case from Previous years.

```python
reported = corruption.sort_values("Cases Reported during the year", ascending = False)
```

```python
plt.figure(figsize=(20,12))
plt.xticks(rotation=90)
sns.pointplot(x='State/UT ', y='Cases Reported during the year', data=reported, color='red');
```

#### maharastra has the highest case reported during the year.
